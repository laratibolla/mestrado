```python
import os
import math
import numpy as np
import pandas as pd
import statistics as stat

dados = pd.read_csv("exerciciocasa1_Idades.csv")
dados
```

```python
#medium #mode #mean - usando biblioteca statistic

#Coluna Paciente
medianaPaciente = stat.median(dados["Paciente"])
modaPaciente = stat.mode(dados["Paciente"])
mediaPaciente = stat.mean(dados["Paciente"])
print("A mediana de Paciente é: % s." % medianaPaciente)
print("A moda de Paciente é % s." % modaPaciente)
print("A média de Pacientes é % s." % mediaPaciente)

#Coluna partos
medianaPartos = stat.median(dados["partos"])
modaPartos = stat.mode(dados["partos"])
mediaPartos = stat.mean(dados["partos"])
print("A mediana de partos é % s." % medianaPartos)
print("A moda de partos é % s." % modaPartos)
print("A média de partos é % s." % mediaPartos)

#Coluna Idades
medianaIdades = stat.median(dados["Idades"])
modaIdades = stat.mode(dados["Idades"])
mediaIdades = stat.mean(dados["Idades"])
print("A mediana de Idades é % s." % medianaIdades)
print("A moda de Idades é % s." % modaIdades)
print("A média de Idades é % s." % mediaIdades)
```

```python
#sem usar função pronta da bilbioteca

#média Coluna Paciente
N = len(dados)
mediaPaciente = sum(dados["Paciente"])/N
print("A média de Pacientes é: %s." %mediaPaciente)

#média Coluna partos
N = len(dados)
mediaPartos = sum(dados["partos"])/N
print("A média de Partos é: %s." %mediaPartos)

#média Coluna Idades
N = len(dados)
mediaIdades = sum(dados["Idades"])/N
print("A média de Idades é: %s." %mediaIdades)

```

```python
# agora sem usar a função statistic

from collections import Counter

#moda Coluna Paciente
paciente = dados["Paciente"]
N = len(dados)
contagempacientes = Counter(paciente)
dicionarioContagem = dict(contagempacientes) #dicionário de dados contados

modaPacientes = [] # inicia uma lista vazia para guardar os valores de moda
for dado, quantidade in dicionarioContagem.items(): # passa por cada elemento de dicionarioContados
    if quantidade == max(list(contagempacientes.values())):
        modaPacientes.append(dado) # .append(dado) adiciona o valor da variavel dado à lista moda

if len(modaPacientes) == N:
    print("Coluna Paciente: Este conjunto de dados não possui moda")
else:
    moda = modaPacientes
    print("Coluna Paciente: A moda de Paciente é % s." % ', '.join(map(str, moda))) 
    
    
#moda Coluna partos
partos = dados["partos"]
N = len(dados)
contagempartos = Counter(partos)
dicionarioPartos = dict(contagempartos) #dicionário de dados contados

modaPartos = [] # inicia uma lista vazia para guardar os valores de moda
for dado, quantidade in dicionarioPartos.items(): # passa por cada elemento de dicionarioContados
    if quantidade == max(list(contagempartos.values())): 
        modaPartos.append(dado) # .append(dado) adiciona o valor da variavel dado à lista moda
        
if len(modaPartos) == N:
    print("Coluna Partos: Este conjunto de dados não possui moda")
else:
    moda = modaPartos
    print("Coluna Partos: A moda de Partos é % s." %','.join(map(str, moda)))
    
    
# moda Coluna Idades
idades = dados["Idades"]
N = len(dados)
contagemIdades = Counter(idades) #cria um objeto Counter com a quantidade de ocorrencias de cada dado
dicionarioIdades = dict(contagemIdades) #dicionário de dados contados

modaIdades = [] # inicia uma lista vazia para guardar os valores de moda
for dado, quantidade in dicionarioIdades.items(): # passa por cada elemento de dicionarioContados
    if quantidade == max(list(contagemIdades.values())):
        modaIdades.append(dado) # .append(dado) adiciona o valor da variavel dado à lista moda
        
if len(modaIdades) == N:
    print("Coluna Idades: Este conjunto de dados não possui moda")
else:
    moda = modaIdades
    print("Coluna Idades: A moda de Idades é % s." %','.join(map(str, moda))) # .join junta os elementos de um vetor
```

```python
# mediana Coluna Paciente

N = len(dados)
pacientes = sorted(dados["Paciente"])

if N % 2 == 0:
    mediana1 = pacientes[N//2]
    mediana2 = pacientes[N//2+1]
    mediana = (mediana1+mediana2)/2
else:
    mediana = pacientes[(N+1)/2]
print("A mediana de Pacientes é % s." % mediana)


# mediana Coluna partos
N = len(dados)
partos = sorted(dados["partos"])
                      
if N % 2 == 0:
    mediana1 = partos[N//2]
    mediana2 = partos[N//2+1]
    mediana = (mediana1+mediana2)/2
else:
    mediana = partos[(N+1)/2]
print("A mediana de partos é % s." %mediana)

# mediana de Idades

N = len(dados)
idades = sorted(dados["Idades"])

if N % 2 == 0:
    mediana1 = idades[N//2]
    mediana2 = idades[N//2+1]
    mediana = (mediana1+mediana2)/2
else:
    mediana = idades[(N+1)/2]
print("A mediana de Idades é % s." %mediana)
                      
```
