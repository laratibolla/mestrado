### DISTRIBUIÇÃO BINOMIAL

Esta variável segue uma distribuição binomial, uma vez que cumpre todos os requisitos: 
    - O número de testes é finito, sendo neste caso 50 pacientes;
    - Só existem dois resultados possíveis a cada teste, receber a vacina ou receber placebo;
    - Os resultados são independentes uns dos outros. 
<p> A distribuição binomial pode ser calculada através da biblioteca scipy ou recorrendo à fórmula: </p>
$$P(k,n)=\frac{n!}{k!(n-k)!}p^k(1-p)^{n-k}$$ 

```python
%matplotlib inline
import os
import matplotlib.pyplot as plt 
import numpy as np
import math
from scipy.stats import binom  #módulo que possui a função binomial
p=np.array([0.2,0.4,0.5,0.6,0.8])
n=29
def f(k,n,p):
    return binom.pmf(k,n,p)  # A CUMULATIVA É CDF probability mass function
k = np.arange(0,29)
for pval in p:
    plt.plot(k,f(k,n,pval),'-o', label=pval)
    plt.legend()
    plt.title("Distribuição binomial")
    plt.legend()
```

$$P(k,n)=\frac{\lambda^k}{k!}e^{-\lambda}$$

### INDICADORES DE ALCANCE

```python
import os
import numpy as np
import pandas as pd
import statistics as stat
import math

mediaPeso = stat.mean(dados["Peso"])
varianciaPeso1= stat.variance(dados["Peso"])
dpPeso1= stat.stdev(dados["Peso"])
print("A variancia da variável peso = % s." % varianciaPeso1)

i = dados["Peso"] #lista com todos os valores da variável peso
a=(i-mediaPeso)**2
varianciaPeso=sum(a)/(len(dados)-1)
print("A variancia da variável peso = % s." % varianciaPeso)
#o desvio padrão é a raiz quandrada da variancia
dpPeso=math.pow(varianciaPeso,1/2)
print("O desvio padrão da variável peso = % s." % dpPeso)
# o erro padrão calcula-se dividindo o desvio padrão pela raiz quadrada do tamanho amostral
erroPeso=dpPeso/math.pow(len(dados),1/2)
print("O erro padrão da variável peso = % s." % erroPeso)
print(" ")
```
