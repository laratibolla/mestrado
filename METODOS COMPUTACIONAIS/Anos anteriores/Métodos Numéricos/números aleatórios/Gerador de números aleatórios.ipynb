<b>Gerador de números aleatórios</b>

Vamos construir o nosso gerador de números aleatórios a partir de:

$X_{n+1}=(aX_n+c)\mod m$

Utilizando os valores:
$m=2^{31}-1,a=16807,X_0=2^{19}-1$


```python
%matplotlib inline
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
m=2**31-1 #modulo
a=16807 #multiplicador
c=0 #incremento
s=2**19-1 #valor inicial ou semente (é o Xo)
rand=[int(10*s/(2**31-1)+2)]
x=[0]
n=10
y=[3,2,2,3,4,6,2,6,1,2]
for i in range (1,n):
    s = (a*s+c)%m
    rand.append(int(10*s/(2**31-1)+2))  #numero aleatorio entre 2 e 10
    x.append(i)
plt.plot(x,rand,'-o')
plt.plot(x,y,'-o')
```

mas o python tem o seu próprio gerador de números aleatórios...

```python
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
from random import random
rand=[]
x=[]
n=100
for i in range(0,n):
    s=random()
    rand.append(int(6*s)+1)
    x.append(i)
plt.plot(x,rand,'-o')
```
