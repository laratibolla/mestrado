<b> Função cumulativa da Gaussiama e valores z</b>

```python
%matplotlib inline
import os
import numpy as np
import math
from scipy.stats import norm  #módulo que possui a função binomial
import matplotlib.pyplot as plt 
sigma=1
mu=0
def f(x,mu,sigma):
    return norm.cdf(x,mu,sigma)
x = np.arange(-5,5,.001)
plt.plot(x,f(x,mu,sigma),'-', label=[sigma,mu])
plt.legend()
```

```python
%matplotlib inline
import os
import numpy as np
import math
from scipy.stats import norm  #módulo que possui a função binomial
import matplotlib.pyplot as plt 
import pandas as pd
sigma=1
mu=0
def f(x,mu,sigma):
    return norm.cdf(x,mu,sigma)
x = np.arange(-3.5,3.5,.001)
cumulativa = {'z':x,'cdf(z)':f(x,mu,sigma)}
df=pd.DataFrame(data=cumulativa)
pd.set_option('max_rows',7000)
df['z'] = df['z'].map('{:,.4}'.format)
df.set_index('z')
#x1=-0.839
#x2=0.558
#print(f(x1,mu,sigma),f(x2,mu,sigma))
```
