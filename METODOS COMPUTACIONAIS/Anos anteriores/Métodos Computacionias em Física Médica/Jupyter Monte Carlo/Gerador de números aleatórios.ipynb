<b>Gerador de números aleatórios</b>

Vamos construir o nosso gerador de números aleatórios a partir de:

$X_{n+1}=(aX_n+c)\mod m$

Utilizando os valores:
$m=2^{32},a=1664525,c=1013904223$


```python
%matplotlib inline
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
m=2**32
a=1664525
c=1013904223
s=274321
rand=[s/4.1e9]
x=[0]
n=100
for i in range (1,n):
    s = (a*s+c)%m
    rand.append(s/4.1e9)
    x.append(i)
plt.plot(x,rand,'-o')
```

mas o python tem o seu próprio gerador de números aleatórios...

```python
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
from random import random
rand=[]
x=[]
n=100
for i in range(0,n):
    s=random()
    rand.append(s)
    x.append(i)
plt.plot(x,rand,'-o')
```
