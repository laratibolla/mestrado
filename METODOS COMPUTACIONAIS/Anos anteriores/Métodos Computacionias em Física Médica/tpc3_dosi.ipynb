```python
# Python program to implement Runge Kutta method 
#Clculo dos valores de lambda para cada compartimento

%matplotlib inline
import numpy as np     
import math
import pandas as pd
import statistics
import matplotlib
import matplotlib.pyplot as plt

ln2=np.log(2)
lambdaAB=0.3*ln2/6
lambdaAC=0.7*ln2/6
lambdaBD=ln2/1920
lambdaDA=0.8*ln2/(12*24)
lambdaDE=0.2*ln2/288
lambdaC0=ln2/12
lambdaE0=ln2/24
lambdaP=ln2/(24*8.02)

print('lambdaAB =', lambdaAB) 
print('lambdaAC =', lambdaAC) 
print('lambdaBD =', lambdaBD) 
print('lambdaDA =', lambdaDA)
print('lambdaDE =', lambdaDE)
print('lambdaC0 =', lambdaC0)
print('lambdaE0 =', lambdaE0)
print('lambdaP =', lambdaP)

#Neste caso as funcoes X sao os "compartimentos"
#C[0]= A #compartimento A sangue
#C[1] = B
#C[2] = C
#C[3] = D
#C[4] = E 

# A sample differential equation "dC/dt" onde C  o compartimento 

def dydx(t,C): #como varia cada compartimento com o tempo, os valores depois sao calculados com o RK
    
        F=[]
        if (C[0] and C[1])>=-3 and (C[0] and C[1])<=3:
            F.append(-(lambdaAB+lambdaAC+lambdaP)*C[0]+lambdaDA*C[3])
            F.append(-(lambdaBD + lambdaP)*C[1] + lambdaAB*C[0])
            F.append(-(lambdaC0 + lambdaP)*C[2] + lambdaAC*C[0])
            F.append(-(lambdaDA + lambdaDE + lambdaP)*C[3] + lambdaBD*C[1])
            F.append(-(lambdaE0 + lambdaP)*C[4] + lambdaDE*C[3])
        else: 
            F=[0,0,0,0,0]
        return F
  
#Finds value of y for a given x using step size h 
#and initial value y0 at x0.
            
def rungeKutta(t0,F,t,h):
    p=open('Fix.txt','w')
    # Count number of iterations using step size or 
    # step height h 
    n = int((t-t0)/h)  
    # Iterate for number of iterations 
    
    y=F        
    for i in range(1, n + 1): 
    
        k1=[]
        k2=[]
        k3=[]
        k4=[]
        k=[0,0,0,0,0]
        
       
        "Apply Runge Kutta Formulas to find next value of y"
          
        k1 = dydx(t0, y)
      
        
        yy = [y[s]+0.5*k1[s] for s in range(5)]
            
        k2 = dydx(t0 + 0.5 * h, yy)
       
        
        yy2 = [y[s]+0.5*k2[s] for s in range(5)]
            
        k3 = dydx(t0 + 0.5 * h, yy2)
        
        
        yy3 = [y[s]+k3[s] for s in range(5)]
            
        k4 = dydx(t0 + h, yy3) 
        #print(k1,k2,k3,k4)
        for s in range(5):        
            k[s] = (h/6)*(k1[s] + 2 * k2[s] + 2 * k3[s] + k4[s])
            #print(k4[s])
       
            
        for j in range(5):
            # Update next value of y 
            y[j] = y[j] + k[j] 
            
        print(y)
        print(t0)
      
        p.write(str(t0))				#Write in txt the cenas to be read by GnuPlot
        for x in y:
            p.write(' '+str(x)) #escreve os valores do vetor X numa linha 
        p.write('\n') #passa para a linha de baixo
        
        # Update next value of x 
        t0 = t0 + h 
        
    return y
        
print ('The value of y at x is:'), rungeKutta(0, [1,0,0,0,0], 100, 0.01) 
```
