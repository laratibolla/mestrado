Fazendo o código sem utilizar o statistics, 

Para a média, 

```python
import os
import math
import numpy as np
import pandas as pd

dados = pd.read_csv("exerciciocasa1.csv") 
Ntotal =len(dados)                        # número de dados
soma=sum(dados["Idades"])                   #calcular a soma
média=soma/Ntotal           
print("A média da variável Idades é % s" % int(round(média)))
```

Utilizadando as frequências para o cálculo da média,

```python
import os
import math
import numpy as np
import pandas as pd

dados = pd.read_csv("exerciciocasa1.csv") 
Idades = dados["Idades"].value_counts().reset_index(name='Count').rename(columns={'Col1':'Col_value'}) 
Idades.columns=['Idades','contagem']
Ntotal=sum(Idades["contagem"])                # usamos a frequência total
soma=sum(Idades["Idades"]*Idades["contagem"])     #somamos ponderando com frequências
média=soma/Ntotal
print("A média da variável Idades é % s." % int(round(média)))
```

Calculando a mediana,

```python
import os
import math
import numpy as np
import pandas as pd

dados = pd.read_csv("exerciciocasa1.csv") 
Ntotal=len(dados)
Idades=dados["Idades"].sort_values().reset_index(drop=True)            # precisamos de ordenar os dados

if Ntotal % 2 == 0:                                # casos de distribuição par
    medianax=Idades[Ntotal//2-1]
    medianay=Idades[Ntotal//2 ]                  # buscar os dois valores centrais
    mediana= (medianax+medianay)/2
else:
    mediana=Idades[Ntotal//2]
print("A mediana da variável Idades é % s" % mediana)
```

Cálculo para a moda,

```python
import os
import math
import numpy as np
import pandas as pd

dados = pd.read_csv("exerciciocasa1.csv") 
Idades = dados["Idades"].value_counts().reset_index(name='Count').rename(columns={'Col1':'Col_value'}) 
Idades.columns=['Idades','Frequências']
v=[]
maxi=0 
for index, row in Idades.iterrows():                   #vai ver qual é o máximo associado à frequência
    if row["Frequências"] > maxi:                      #se for maior é esse valor que fica
          maxi=row["Frequências"]
for index, row in Idades.iterrows():                   #depois vamos encontrar o valor das idades associadas a essa freq.
    if row["Frequências"] == maxi:
          v.append(row["Idades"])

print("As modas da variável Idades são % s." % v)
```

Comparando os valores do excel, com os valores obtidos no phyton para a média,moda e mediana, obtiveram-se resultados satisfatórios.
Agora, como o Python permite também determinar valores de localização não central - os quantis, vamos determinar os quantis, percentis e decis. Utilizando o "statistics".

```python
import os
import math
import numpy as np
import pandas as pd
import statistics

percent=[0]
n=0.00
for i in range (1,100):
    n=n+0.01
    percent.append(n)
    
dados = pd.read_csv("exerciciocasa1.csv") 
percentis = dados["Idades"].quantile(percent).reset_index()         #podemos arredondar
percentis.columns=['percentil','valor']
print("Valores para os percentis das Idades, ")
percentis
```
