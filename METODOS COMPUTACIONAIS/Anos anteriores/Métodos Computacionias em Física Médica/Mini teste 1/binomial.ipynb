```python
%matplotlib inline
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd

from scipy.stats import binom  #módulo que possui a função binomial
def factorial(x):
    factorial=1
    for i in range(1,x+1):
        factorial = factorial*i
    return factorial
def binomial(s, p, n):
    binomial = factorial(n) / (factorial(s)*factorial(n-s))* p**s * (1 - p)**(n-s)
    return binomial
def binom(s,n,p):
    binom=[]
    for i in s:
        binom.append(binomial(i,n,p))
    return binom
def cumulative(s, p, n):
    cum = 0.0
    a = n - s
    c = n
    for j in range(0,s+1):
        cum += factorial(c) / (factorial(j)*factorial(c-j))* p**j * (1 - p)**(c-j)
    return cum
def cumul(s,n,p):
    cumul=[]
    for i in s:
        cumul.append(cumulative(i,n,p))
    return cumul
p=np.array([0.2,0.4,0.5,0.6,0.8])     # p representa as várias probabilidades
n=20                                  # n representa o número de tentativas
mode=[]
median=[]
mean=[]
variance=[]
stdev=[]
for pval in p:                       #para a binomial, a média é dada por np, sendo pval a probabilidade 
    mean1=n*pval                     #de um determinado valor , val ; mean.append-> vai alocar o valor
    mean.append(mean1)
    variance1=n*pval*(1-pval)               # variância é dada por npq sendo q=1-p ;
    variance.append(variance1)
    stdev1=round(math.sqrt(variance1),2)    # o desvio padrão é dado pela raiz quadrada da variância
    stdev.append(stdev1)
    if n%2==0:                             # se o nr de tentativas(n) for par,  
        median1=math.floor((n+1)*pval)     # a mediana é (n+1)p
        mode1=math.floor((n+1)*pval)       # a moda é (n+1)p
    else:
        median1=math.floor(n*pval)        # se o nr de tentativas(n) for ímpar
        mode1=math.floor((n+1)*pval-1)    #a mediana é np ; a moda é (n+1)p-1
    median.append(median1)          
    mode.append(mode1)
S=len(p)
dados = {'probabilidade': p, 'média': mean,'mediana': median,'moda': mode,"variância": variance,"desvio padrão": stdev}
df = pd.DataFrame(data=dados).set_index('probabilidade')
df
```
