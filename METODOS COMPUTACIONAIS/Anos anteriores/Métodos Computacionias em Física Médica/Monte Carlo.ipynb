**É possível fazer o cálculo de integrais utilizando amostras de números aleatórios.** 
<ul>
<li>Cálculo do integral  $f(x)=1/(x^2-1)$, $0\leq x\leq 10$ , utilizando as técnicas de Monte Carlo</li>. 
    
Os números possíveis de obter, são usualmente desingados como sendo aleaórios. No entanto números aleatórios são números uniformemente distribuídos ao longo de um intervalo definido enão existe forma d eprever qual vai ser o valor futuro numa sequência. Um computador é determinista e apesar de serem designados como sendo aleatórios, os números calculados são na realidade pseudo-aleatórios.
   
Para calcular o integral utilizando números aleatórios, basta pensar que o cálculo de um integral, corresponde ao cálculo da área por baixo da função  $f(x)$. Para isso, como primeira aporximação,o programa vai correr dois números aleatórios $\xi_a$ e $\xi_b$ de 0 até 10.  

```python
# Para determinar dois números aleatórios, é simples utilizando o random()

%matplotlib inline
from random import seed
from random import random
import os
import os
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

seed(1)
xa=10.0*random()     # valor de 0 até 10
xb=10.0*random()

print("Número aleatório é " ,xa)
print("Número aleatório é " ,xb)
```

Já estão determinados os números aleatórios, 

Agora, é preciso determinar $f(\xi_a)$ e $f(\xi_b)$ e calcular o valor da área do rectângulo definido por cada um desses valores com altura $h_1=f(\xi_a)$ e $h_2=f(\xi_b)$, tendo de largura de 0 até 10 ($l=10-0$) que corresponde a $l=10$.

Área 1 -> $A_{\xi_a}=l*h=(10-0)f($\xi_a$)$;

Área 2 -> $A_{\xi_b}=l*h=(10-0)f($\xi_b$)$; 

-------------------------------------------------------------------------------

Agora, ao fazer uma média das duas áreas, obtemos uma primeira estimativa do integral.

$\displaystyle I\approx \frac{1}{2}(A_{\xi_a}+A_{\xi_b})=\frac{1}{2}(10-0)[f({\xi_a})+f({\xi_b})]$

Esta aproximação será relativamente fraca, mas à medida que é feito o cálculo da média, infinitas vezes, o valor calculado irá tender para o valor do integral.

```python
#Agora tendo em conta os números determinados acima, será feito o cálculo da área abaixo desses números,
#Pretendemos obter a média das áreas obtidas(graficamente) sendo isso aproximadamente igual a intgeral que se pretende obter. 

from random import seed
from random import random
import os
import os
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

xa=1.3436424411240122
xb=8.4743373693723270
# A função y(x) é dada por y(x)=1/(x^2 + 1)
ya=[1/(xa**2 +1)]
yb=[1/(xb**2 +1)]

x= np.arange(0., 10., 0.01)  
for i in range(0,10,1):           #i de 0 até 10
    ya.append(1/(xa**2 +1))       #vai alocar os valores
    yb.append(1/(xb**2 +1))
xa = np.arange(len(ya))
xb = np.arange(len(yb))           #para o calculo da área ser feito com comp.* altura

#Plot dos gráficos;
fig, (axa, axb) = plt.subplots(1,2,sharex=True,figsize=(10, 10)) #sharex=True,x- or y-axis will be shared among all subplots.
           
axa.plot(x,1/(x**2+1),color='black')   
axa.fill_between(xa, 0, ya,color='blue')    #serve para preencher a area   
axa.plot(xa, ya, 'pink')
#para dar nome aos eixos do plot
axa.set_ylabel('f(x)')         
axa.set_xlabel('x')

axb.plot(x,1/(x**2+1),color='orange')
axb.fill_between(xb, 0, yb,color='black')    #serve para preencher a area 
axb.plot(xb, yb, 'yellow')
#para dar nome aos eixos do plot
axb.set_ylabel('f(x)')
axb.set_xlabel('x')

```

No entanto, como foi referido acima, esta aproximação é muito fraca. Devido ao teroema do limite central, à medida que  n tende para infinito, o valor calculado irá tender para o valor do integral.


<ul>
<li>Agora, para obtermos um resultado mais fiável, o cálculo do integral pode ser feito recorrendo ao Monte Carlo;</li>. 


```python
#Cálculo do integral y(x)=1/(x^2 + 1), 0<=x<=10 

%matplotlib inline
from random import seed
from random import random
import os
import os
import math
from math import pi
from math import sqrt
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

seed(1)
a2=[0]
b2=[0]
res=[np.arctan(10)]   
for Ntotal in range (1,50000,10):        #(1,N,p) quanto mais se aproximar N do infinito, melhor será o valor obtido.  
    x=0.00000000                         #(1,N,p) os valores são simulados de p em p pontos
    a=np.arctan(10)
    total=0.00000
    for i in range(Ntotal):          #simular N números aleatórios e somar
        r=10.0*random()
        x = x + 1/((r**2) + 1)     
        
    float(total)
    total = (10.0/Ntotal)*x         #média dos valores simulados
    a2.append(Ntotal)
    b2.append(total)
    res.append(a)

print('Valor do integral é',a)
print('Valor calculado pelo método de monte carlo é',total)
plt.plot(a2, b2,':r')
plt.plot(a2, res,'--',color='blue')
```
